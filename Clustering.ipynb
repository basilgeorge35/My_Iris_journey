##Task 2.1

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
label = np.load('Iris_labels.npy')
```

```python
print("The shape of the Iris matrix is ", label.shape)
```

```python
data = np.load('Iris_data.npy')
```

```python
plt.figure()
plt.scatter(data[:,0], data[:,1], c=label)
plt.title("Iris Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

##Task 2.2

```python
#Initializing scikit-learn KMeans object

from sklearn.cluster import KMeans
```

```python
#Fitting k-means model to Iris_data

kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(data)
```

```python
#Predicting cluster memberships
kmeans.fit_predict(data)


# kmeans.labels_
# kmeans.cluster_centers_
```

```python
#Plotting K cluster Model with cluster centers

plt.figure()
plt.scatter(data[:,0], data[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0] , kmeans.cluster_centers_[:,1] , s = 70, marker="x", color = 'r',linewidths=3)
plt.title("KMeans Iris Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```

```python
#We can create a better clustering model by increasing number of clusters to 4

kmeans = KMeans(n_clusters=4, n_init='auto')  #n_clusters increased to 4
kmeans.fit(data)

plt.figure()
plt.scatter(data[:,0], data[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0] , kmeans.cluster_centers_[:,1] , s = 70, marker="x", color = 'r',linewidths=3)
plt.title("KMeans Iris Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()


# init : {â€˜k-means++â€™, â€˜randomâ€™} -- Dictates how initial centroids are selected
# n_init : â€˜autoâ€™ or int -- Dictates the number of times the k-means algorithm is run with different centroid seeds.
```

##Task 2.3

```python
#initializing instance of scikit-learn GaussianMatrix

from sklearn.mixture import GaussianMixture
```

```python
#Fitting Iris_data to the model

gm = GaussianMixture(n_components=3, random_state=0).fit(data)
```

```python
#Predicting clusters for Iris_data and assigning it to labels array
labels = gm.predict(data)

#Plotting the graph
plt.figure()
plt.scatter(data[:,0], data[:,1], c=labels)

#Adding gaussian mean to the plot
plt.scatter(gm.means_[:,0] , gm.means_[:,1] , s = 70, marker="x", color = 'r', linewidths=3)
plt.title("Gaussian Mixture Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
np.shape(gm.predict_proba(data))

# predict_proba() gives probability density of each data point vs each cluster

#x = np.amax(gm.predict_proba(data), axis=1)
#print(x)
#labels
```

```python
plt.figure(figsize=(20,5))
posterior_prob = gm.predict_proba(data)

for i in range(3):
  plt.subplot(1, 3, (i+1))

  plt.title(f"Component {i}")
  plt.xlabel("Sepal Length")
  plt.ylabel("Sepal Width")
  plt.scatter(data[:,0], data[:,1], c=posterior_prob[:,i], s=50, cmap='viridis')
  plt.colorbar()
  plt.tight_layout()
```
